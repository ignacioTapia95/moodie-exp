```python
import numpy as np
from PIL import Image
import requests
import matplotlib.pyplot as plt

from transformers import CLIPProcessor, CLIPModel
```

```python
model = CLIPModel.from_pretrained("openai/clip-vit-large-patch14")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-large-patch14")
```

```python
shot_types = [
    "Extreme Long Shot",  # Plano general extremo
    "Long Shot",          # Plano general
    "Full Shot",          # Plano entero
    "Medium Long Shot",   # Plano americano
    "Medium Shot",        # Plano medio
    "Medium Close-Up",    # Plano medio corto
    "Close-Up",           # Primer plano
    "Extreme Close-Up",   # Primerisimo primer plano
    "Insert Shot",
]
```

```python
shot_types = [
    "Gran Plano General",  # Plano general extremo
    "Plano General",          # Plano general
    "Plano Americano",   # Plano americano
    "Plano Medio",        # Plano medio
    "Primer Plano",           # Primer plano
    "Primer√≠simo Primer Plano",   # Primerisimo primer plano
    "Inserto",
]
```

```python
path = "../../data/processed/image/tt4652838/tt4652838-Scene-828-02.jpg"
image = Image.open(path)

inputs = processor(text=shot_types, images=image, return_tensors="pt", padding=True)
outputs = model(**inputs)
logits_per_image = outputs.logits_per_image
probs = logits_per_image.softmax(dim=1)

predicted_index = np.argmax(probs.cpu().detach().numpy())
predicted_label = shot_types[predicted_index]
predicted_prob = probs[0, predicted_index].item()

plt.imshow(image)
plt.axis("off")
plt.title(f"Predicci√≥n: {predicted_label}\nProbabilidad: {predicted_prob:.2f}")
plt.show()
```

### Angulo de Camara

```python
camera_angles = [
    "High Angle",           # √Ångulo picado (c√°mara por encima del sujeto)
    "Low Angle",            # √Ångulo contrapicado (c√°mara por debajo del sujeto)
    "Eye Level",            # √Ångulo a nivel de los ojos (neutral y natural)
    "Extreme High Angle",   # √Ångulo picado extremo (c√°mara muy por encima del sujeto)
    "Extreme Low Angle",    # √Ångulo contrapicado extremo (c√°mara muy por debajo del sujeto)
    "Bird's Eye View",      # √Ångulo cenital (vista desde directamente arriba)
    "Worm's Eye View"       # √Ångulo nadir o de gusano (vista desde directamente abajo)
]
```

```python
path = "../../data/processed/image/tt4652838/tt4652838-Scene-483-02.jpg"
image = Image.open(path)

inputs = processor(text=camera_angles, images=image, return_tensors="pt", padding=True)
outputs = model(**inputs)
logits_per_image = outputs.logits_per_image
probs = logits_per_image.softmax(dim=1)

predicted_index = np.argmax(probs.cpu().detach().numpy())
predicted_label = camera_angles[predicted_index]
predicted_prob = probs[0, predicted_index].item()

plt.imshow(image)
plt.axis("off")
plt.title(f"Predicci√≥n: {predicted_label}\nProbabilidad: {predicted_prob:.2f}")
plt.show()
```
